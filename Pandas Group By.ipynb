```python
# Bonus: convert MPG columns to km_per_liter

MPG = Miles/Gallon -> Km/Liter

1 Mile = 1.60934 Km

1 Gallon = 3.78541 Liters

$$ \frac{Miles}{Gallon} -&gt; \frac{Miles}{Gallon} * \frac{Km}{Miles} * \frac{Gallon}{Liters}$$$$ \frac{Miles}{Gallon} -&gt; \frac{Miles}{Gallon} * \frac{1.60934Km}{ 1Miles} * \frac{1 Gallon}{3.78541 Liters}$$
( 1.60934 / 3.78541 )
```

```python
cars['City Km/Liter'] = list( map(lambda x: x * ( 1.60934 / 3.78541 ),cars['City MPG']) )
```
